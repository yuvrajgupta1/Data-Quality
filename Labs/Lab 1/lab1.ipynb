### Lists, Loops, and Conditionals

```python
my_list = [1,2,3,4,5,6]

for item in my_list:
    if item % 2 == 0:
        print(item)
```

Here is another way of doing the same thing, this time using explicit list indexing.

```python
for i in range(len(my_list)):
    item = my_list[i]
    if item % 2 == 0:
        print(item)
```

Or we could write a function that encapsulates this set of instructions:

```python
def print_evens(some_list):
    for item in some_list:
        if item % 2 == 0:
            print(item)

a_list = [10,11,12,13,14,15]
print_evens(a_list)
```

# Working with Data in Python

```python
import pandas as pd
```

# Reading in Data

We'll read in a dataset containing some information about cars.

It's in a CSV file.

```python
'''
If you're running Jupyter in a different directory from the CSV file, 
then you need to supply the path to the CSV file below.
'''

Location = 'mtcars.csv'


df = pd.read_csv(Location)
```

# Examining Data

In Pandas, we call this data structure a Data Frame.

You can think of it as being similar to an Excel spreadsheet. 

```python
df
```

We can use the head() function of the data frame to look at the first 5 records.

```python
df.head()
```

Similarly, we can use the tail() function to look at the last 5 records.

```python
df.tail()
```

## Summarizing a Data Frame

We can use the describe() function to get a summary of a data frame.


```python
df.describe()
```

## Indexing Data Frames

We can index specific columns of the data frame.

```python
df['model']
```

```python
df['mpg']
```

We may want to get rows of the data frame that have particular values for a column. 

Notice the following statement gives us Boolean values. Each car either gets more than 20 miles per gallon, or it doesn't. 

```python
df['mpg'] > 20
```

We can use those Boolean values as an index into the data frame

```python
high_mpg = df[df['mpg'] > 20]
print(high_mpg)
```

Another example. Get just the cars with four cylinders:

```python
four_cyl = df[df['cyl'] == 4]
print(four_cyl)
```

## Advanced Indexing

If we want to specify row and column indices, we can use the _iloc_ attribute

```python
print(df.head())

# indexing starts at 0,
# so this will print out the third row, third column value
print("\nnum. cylinders of third car: ", df.iloc[2,2])
```

The following statement says to print out all of the columns for the fourth and fifth rows.

```python
print(df.iloc[3:5,:])
```

The following statement says to print out all the rows for columns five and six.

```python
print(df.iloc[:,4:6])
```
